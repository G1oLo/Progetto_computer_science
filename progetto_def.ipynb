## 3. For each (primary) breed, determine the ratio between the number of dogs of Mixed Breed and those not of Mixed Breed. Hint: look at the secondary_breed.

```python
dogs['breed_secondary'].unique()
```

Primo approccio: considerare mixed breed i cani che presentano come razza secondaria all'interno del loro DNA 'Mixed Breed'

```python
dogs_mixed = dogs[dogs['breed_secondary'] == 'Mixed Breed']
dogs_not_mixed = dogs[dogs['breed_secondary'] != 'Mixed Breed']

grouped_mixed = dogs_mixed.groupby('breed_primary').size().reset_index(name='mixed_counts')
grouped_not_mixed = dogs_not_mixed.groupby('breed_primary').size().reset_index(name='not_mixed_counts')

result = grouped_mixed.merge(grouped_not_mixed, how='outer' ,on='breed_primary')
result['mixed_counts'] = result['mixed_counts'].fillna(0)
result['not_mixed_counts'] = result['not_mixed_counts'].fillna(0) 

result['ratio'] = result['mixed_counts'] / result['not_mixed_counts']
result
```

Secondo approccio: considerare mixed breed i cani che presentano valore 'True' nella colonna breed_mixed, ovvero i cani che presentano all'interno del loro DNA una razza secondaria diversa da quella primaria

```python
dogs_mixed2 = dogs[dogs['breed_mixed']== True]
dogs_not_mixed2 = dogs[dogs['breed_mixed']!= True]

grouped_mixed2 = dogs_mixed2.groupby('breed_primary').size().reset_index(name='mixed_counts')
grouped_not_mixed2 = dogs_not_mixed2.groupby('breed_primary').size().reset_index(name='not_mixed_counts')

result2 = grouped_mixed2.merge(grouped_not_mixed2, how='outer' , on='breed_primary')

result2['mixed_counts'] = result2['mixed_counts'].fillna(0)
result2['not_mixed_counts'] = result2['not_mixed_counts'].fillna(0) 

result2['ratio'] = result2['mixed_counts'] / result2['not_mixed_counts']
#dogs_mixed2["new_ratio"].replace([np.inf, -np.inf], 0, inplace=True)  
result2
```
